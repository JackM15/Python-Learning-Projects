```python
#import pandas
import pandas
#import geopy lat/long shizzle
from geopy.geocoders import Nominatim

#Set the nominatim to the nom class
nom = Nominatim()
```

```python
#Open up the excel file
datfile = pandas.read_excel("Xmo Strata Ltd Analysis 2014 - May.xls", 1)
#show the dataframe
datfile
```

```python
#Select the correct column with the data in it with addresses
column18 = datfile[datfile.columns[18:19]]
column18
```

```python
#Filter out the NaN entries
column18 = column18[pandas.notnull(column18['Unnamed: 18'])]
column18
```

```python
#Create a column named coordinates based on the data from unnamed 18. then apply the nom.geocode method to each.
column18["Coordinates"] = column18["Unnamed: 18"].apply(nom.geocode)
column18
```
